```python
import numpy as np
data_num = data.select_dtypes(np.number)
data_num
```

```python
def remove_outliner(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3-q1
    df_final = df[(df > (q1 - 1.5*IQR)) | (df < (q3 + 1.5*IQR))]
    return df_final
```

```python
new_data = data_num.apply(remove_outliner)
```
